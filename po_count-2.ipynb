```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
data_fest = pd.read_csv('/Users/iraraygorod/Downloads/5024.csv', sep = ';')
data_fest['Наименование ОУ'] = data_fest['Наименование ОУ'].str.lower()
data_fest['Наименование ОУ'] = data_fest['Наименование ОУ'].str.replace('"', '', regex = True).replace('»', '', regex = True).replace('«', '', regex = True)
```

```python
data_fest.columns = ['id', 'date', 'lastname', 'firstname', 'secondname', 'birthdate', 'sex', 'phone', 'email', 'fo', 'ro', 'mo', 'school', 'course', 'region', 'gr', 'role', 'status', 'rolepl', 'dateot', 'regvar', 'ref', 'okrug']
```

```python
schools_fest = pd.pivot_table(data_fest, index = 'school', values = 'id', aggfunc = 'count').reset_index()
schools_fest
```

```python
data_conf = pd.read_csv('/Users/iraraygorod/Downloads/5025.csv', sep = ';')
data_conf['Наименование ОУ'] = data_conf['Наименование ОУ'].str.lower()
data_conf['Наименование ОУ'] = data_conf['Наименование ОУ'].str.replace('"', '', regex = True).replace('»', '', regex = True).replace('«', '', regex = True)
```

```python
data_conf.columns = ['id', 'date', 'lastname', 'firstname', 'secondname', 'birthdate', 'sex', 'phone', 'email', 'fo', 'ro', 'mo', 'school', 'course', 'region', 'gr', 'role', 'status', 'rolepl', 'dateot', 'regvar', 'ref', 'okrug']
```

```python
schools_conf = pd.pivot_table(data_conf, index = 'school', values = 'id', aggfunc = 'count').reset_index()
schools_conf
```

```python
school_data = pd.read_csv('/Users/iraraygorod/Downloads/Школы.csv')
school_data.columns = ['school']
school_data['school'] = school_data['school'].str.lower().replace('"', '', regex = True).replace('»', '', regex = True).replace('«', '', regex = True)
school_data
```

```python
good_data = school_data.merge(schools_fest, how = 'left', left_on='school', right_on='school').merge(schools_conf, how = 'left', left_on='school', right_on='school')
good_data = good_data.fillna(0)
good_data.columns = ['school', 'fest', 'conf']
good_data['fest'] = good_data['fest'].astype('int32')
good_data['conf'] = good_data['conf'].astype('int32')
good_data.sort_values('conf', ascending = False)
```

```python
good_data.to_csv('/Users/iraraygorod/Downloads/schools_number.csv')
```

```python
data_orders = pd.read_csv('/Users/iraraygorod/Downloads/orders_г_Москва_all (4).csv', sep = ';')
data_orders['Учебное заведение кандидата'] = data_orders['Учебное заведение кандидата'].str.lower().replace('"', '', regex = True).replace('»', '', regex = True).replace('«', '', regex = True)
data_orders['Первичное отделение'] = data_orders['Первичное отделение'].str.lower().replace('"', '', regex = True).replace('»', '', regex = True).replace('«', '', regex = True)
```

```python
school_orders = pd.pivot_table(data_orders, index = 'Учебное заведение кандидата', values = 'ФИО пользователя', aggfunc = 'count').reset_index()
#school_orders['Учебное заведение кандидата'] = school_orders['Учебное заведение кандидата'].str.lower()
school_orders.columns = ['school', 'number_study'] 
school_orders.sort_values('number_study', ascending = False)
```

```python
school_po_orders = pd.pivot_table(data_orders, index = 'Первичное отделение', values = 'ФИО пользователя', aggfunc = 'count').reset_index()
school_po_orders.columns = ['school', 'number_po'] 
school_po_orders.sort_values('number_po', ascending = False)
```

```python
school_full = school_orders.merge(school_po_orders, how = 'outer')
school_full
```

```python
good_data_full = good_data.merge(school_full, how = 'left')
good_data_full
```

```python
good_data_full = good_data_full.fillna(0)
good_data_full.head()
```

```python
good_data_full['number_study'] = good_data_full['number_study'].astype('int32')
good_data_full['number_po'] = good_data_full['number_po'].astype('int32')
```
